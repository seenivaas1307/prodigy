***Importing Necessary Packages***

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/train (1).csv')
```

***Fill missing values***

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop(columns=['Cabin'], inplace=True)
```

***DROPPING UNNECESSARY COLUMNS AND CONVERTING CATEGORICAL TO NUMERIC FEATURE***

```python
df.drop(columns=['Ticket', 'Name'], inplace=True)
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)
```

***Exploratory Data Analysis (EDA)***

***1. Survival count by gender ***

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex_male', hue='Survived', data=df, palette='Set2')
plt.title('Survival Count by Gender')
plt.show()
```

***2.Survival count by passenger class***

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=df, palette='Set1')
plt.title('Survival Count by Passenger Class')
plt.show()
```

***3. Age distribution by survival status***

```python

plt.figure(figsize=(8, 6))
sns.histplot(df[df['Survived'] == 1]['Age'], bins=30, kde=False, color='green', label='Survived', alpha=0.6)
sns.histplot(df[df['Survived'] == 0]['Age'], bins=30, kde=False, color='red', label='Not Survived', alpha=0.6)
plt.title('Age Distribution by Survival Status')
plt.legend()
plt.show()
```

***4. Survival rate based on fare paid***

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare Paid by Survival Status')
plt.show()
```

***5. Correlation Heatmap (Now works with numerical data only)***

```python

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

***6. Survival rate based on SibSp (Siblings/Spouses aboard)***

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='SibSp', hue='Survived', data=df, palette='Set3')
plt.title('Survival Count by Number of Siblings/Spouses Aboard')
plt.show()
```

***7. Survival rate based on Parch (Parents/Children aboard)***

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Parch', hue='Survived', data=df, palette='Set3')
plt.title('Survival Count by Number of Parents/Children Aboard')
plt.show()
```

***8. Pairplot of numerical variables***

```python
sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare', 'SibSp', 'Parch']], hue='Survived', palette='Set1')
plt.title('Pairplot of Numerical Variables by Survival Status')
plt.show()
```

***9. Survival rate based on embarkation point***

```python

plt.figure(figsize=(8, 6))
sns.countplot(x='Embarked_Q', hue='Survived', data=df, palette='Set2')
plt.title('Survival Count by Embarkation Point')
plt.show()
```

***10. Fare distribution by class***

```python

plt.figure(figsize=(8, 6))
sns.violinplot(x='Pclass', y='Fare', data=df, palette='Set2')
plt.title('Fare Distribution by Passenger Class')
plt.show()
```
